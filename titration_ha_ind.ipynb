# Titration einer einprotonigen Säure mit Indikator

```python
from matplotlib import pyplot as plt
from matplotlib import style
from scipy import stats
from scipy import constants as cnst
import numpy as np
import pandas as pd
import glob
import math
from IPython.display import HTML
```

```python
pKs = np.double(input("pKs der Säure: "))
pKi = np.double(input("pKs des Indikators: "))

# Konstanten
Ks = math.pow(10,-pKs)
Ki = math.pow(10,-pKi)
Kb = 1000
Kw = np.longdouble(1e-14)

# vorgelegtes Volumen und Konzentration der Säure
v0 = 0.02
cs = 0.1

# Konzentration des Indikators
ci = 0.001

# Konzentration der Base
cb = 0.1

# Startwerte
h = 0.1
hi = ci
ha = 0.1
moh = 0.1

# Bürette füllen und Tropfengröße
vol = np.arange(0.0,0.04,0.0002)

pH = []

for v in vol:
    while True:
        while True:
            while True:
                while True:
                    c0s = cs*v0/(v0+v)
                    c0b = cb*v/(v0+v)
                    c0i = ci * v0/(v0+v)
                    a = Ks*ha/h
                    oh = Kw/h
                    i = Ki * hi/h
                    m = Kb*moh/oh
                    if (v == 0):
                        m = 0
                    sk = h + m # Ladungsgleichheit
                    sa = a + oh + i
                    if(abs(sk/sa-1)<1e-3):
                        break
                    h = h*(sa/sk)**0.5
                x1 = ha + a # Errechnete Konz der Säure muss der vorgegebenen  Konz. c0s entsprechen
                if(abs(x1/c0s-1)<1e-3):
                    break
                ha = ha * (c0s/x1)**0.5
                if(v == 0):
                    break
            if(v == 0):
                break
            x2 = m + moh # Errechnete Konz der Base muss der vorgegebenen Konz. c0b entsprechen
            if(abs(x2/c0b-1)<1e-3):
                break
            moh = moh * (c0b/x2)**0.5
        if(v == 0):
            break
        x3 = hi + i
        if(abs(x3/c0i-1)<1e-3):
            break
        hi = hi * (c0i/x3)**0.5
    pH.append(-math.log10(h))

# Darstellung der Daten
x = 1000*vol
fig = plt.figure(figsize=(12, 10)) 
plt.minorticks_on()
plt.grid(True,which='both')
plt.xlabel('V(Base) im ml')
plt.ylabel('pH')
plt.title('Titration einprotonige Säure')
plt.plot(x,pH)
plt.show()
fig.savefig('test.png')

# Daten in Tabellenform
df_pH = pd.DataFrame(data=pH, columns=['pH'])
df_vol = pd.DataFrame(data=x,columns=['ml NaOH'])
df = pd.concat([df_vol,df_pH],axis=1)
display(HTML(df.round(2).to_html(index=True)))
df.to_csv('Test.csv')
```
